```python
#define node for circular linked list
class Node(object):
  def __init__(self):
    self.data=None
    self.next = None

class CircularLinkedList(object):
  def __init__(self,node = None):
    self.head =None
    self.length = 0

  def insertAtFirst(self,data):
    newNode = Node()
    newNode.data = data
    if self.length ==0:
      newNode.next = newNode
      self.head = newNode
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      newNode.next = self.head
      current.next = newNode
      self.head = newNode
    self.length +=1

  def insertAtEnd(self,data):
    newNode = Node()
    newNode.data =data
    if self.length ==0:
      newNode.next = newNode
      self.head = newNode
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      newNode.next = self.head
      current.next = newNode
    self.length +=1

  def insertAtPosition(self,data, position):
    newNode = Node()
    newNode.data =data
    if self.length ==0:
      print("linked list is empty")
      return
    else:
      if position>self.length or position<0:
        print("invalid position")
        return
      else:
        current=self.head
        count =0
        while current.next !=self.head:
          if count+2 == position:
            break
          count +=1
          current =current.next
        newNode.next = current.next
        current.next = newNode
        self.length +=1

  def deleteLastItem(self):
    if self.length ==0:
      print("linked list is empty")
      return
    else:
      current = self.head
      pre=None
      while current.next != self.head:
        pre=current
        current = current.next
      pre.next=current.next
      self.length-=1

  def deleteFirstItem(self):
    if self.length ==0:
      print("linked list is empty")
      return
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      self.head = self.head.next
      current.next=self.head
      self.length-=1

  def deleteAtPosition(self,position):
    if self.length ==0:
      print("linked list is empty")
      return
    else:
      if position>self.length or position<0:
        print("invalid position")
        return
      elif position ==1:
        self.deleteFirstItem()
      else:
        current = self.head
        count =0
        pre = None
        while current.next != self.head:
          if count +1== position:
            break
          count+=1
          pre = current
          current = current.next
        pre.next = current.next
        self.length -=1


  def readLinkedList(self):
    if self.length ==0:
      print("linked list is emtpy")
      return
    else:
      current = self.head
      while current.next != self.head:
        print (current.data)
        current = current.next
      print(current.data)

```
